```sql
select * 
from `workspace`.`default`.`employees`;


--------1. COUNT() FUNCTION-----
select COUNT(*) AS TOTAL_NUMBER_OF_EMPLOYEES
from `workspace`.`default`.`employees`;


-------2. SUM() FUNCTION
SELECT SUM(salary) AS TOTAL_SALARY
from `workspace`.`default`.`employees`;


---- 3. AVE() FUNCTION------
SELECT AVG(salary) AS average_salary
from `workspace`.`default`.`employees`
where department = 'HR';


----4. MIN() AND MAX() FUNCTION-------

SELECT MIN(salary) AS LOWEST_SALARY,
       MAX(salary) AS HIGHEST_SALARY
from `workspace`.`default`.`employees`;


-----5. GROUP BY STATEMENT---------

SELECT department,
  SUM(salary) as total_salary
from `workspace`.`default`.`employees`
GROUP BY department;


---- 6. GROUP BY AND ORDER BY------

SELECT COUNT(*) AS TOTA_EMPLOYEES,
      city
from `workspace`.`default`.`employees`
GROUP BY city;


------7. GROUP BY AND ORDER BY-----
SELECT AVG(salary) AS ave_salary,
       department
from `workspace`.`default`.`employees`
GROUP BY department
ORDER BY ave_salary DESC;

---- 8. HAVING CLAUSE-----

SELECT department,
      SUM(salary) AS total_salary
from `workspace`.`default`.`employees`
GROUP BY department
HAVING total_salary > '100000';

----9. COMBINING GROUP BY, HAVING AND ORDER BY------

SELECT COUNT(*) AS NO_OF_EMPLOYEES,
       city
from `workspace`.`default`.`employees`
GROUP BY city 
HAVING NO_OF_EMPLOYEES > '1'
ORDER BY NO_OF_EMPLOYEES DESC;

----10. COMBINING AGGREGATE FUNCTIONS

SELECT AVG(salary) as AVG_SALARY,
       department
from `workspace`.`default`.`employees`
GROUP BY department
ORDER BY AVG_SALARY
LIMIT 1;     
```
